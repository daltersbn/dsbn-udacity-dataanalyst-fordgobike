## Introduction

This dataset contains bike-sharing trip data. The goal is to analyze usage patterns and user behavior.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

df = pd.read_csv('201902-fordgobike-tripdata.csv')
df.head()
```

## Data Wrangling

```
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

df['start_hour'] = df['start_time'].dt.hour
df['start_day'] = df['start_time'].dt.day_name()
df['duration_min'] = df['duration_sec'] / 60

df['age'] = 2019 - df['member_birth_year']

df = df[df['age'] < 100]
df.dropna(inplace=True)
```

## Trip Duration Distribution

```
sns.histplot(df['duration_min'], bins=50)
plt.xlim(0,60)
plt.show()
```

## Trips by Hour

```
sns.countplot(x='start_hour', data=df)
plt.show()
```

## Trips by Day

```
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
sns.countplot(x='start_day', data=df, order=order)
plt.xticks(rotation=45)
plt.show()
```

## User Type

```
sns.countplot(x='user_type', data=df)
plt.show()
```

## Duration by User Type

```
sns.boxplot(x='user_type', y='duration_min', data=df)
plt.ylim(0,60)
plt.show()
```

## Conclusion

- Trips are short
- Peak hours match commuting
- Subscribers dominate
- Customers ride longer
